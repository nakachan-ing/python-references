## データを並び替える　| sort_values

`sort_values`メソッドを使用するとデフォルトで「昇順」に並び替えることができます。

引数`by`に並び替えしたいカラムを指定します。

### DataFrameの１つのカラムのデータを並び替える

平成27年のデータを抽出し、人口（総数）の少ない順（昇順）に並び替えてみます。

````python
df_h27 = df[(df['元号'] == '平成') & (df['和暦（年）'] == 27.0)]
df_h27
# df[df['西暦（年）'] == 2015.0]でも同じ結果
````

````python
df_h27.sort_values(by='人口（総数）', ascending=True).head(10)
````

**ポイント**

- データを降順に並び替える場合は、引数`by`に`False`を指定します。
- 引数`by`のデフォルトは`True`（昇順）です。

````python
df_h27.sort_values(by='人口（総数）', ascending=False).head(10)
````

### DataFrameの２つ以上のカラムのデータを並び替える

２つ以上のカラムを並び替えたい場合は、引数`by`にリストで指定します。

年号が「大正」のデータを抽出し、「都道府県」と「人口（総数）」をリストで指定して並び替えて見ます。

````python
df_taisho = df[df['元号'] == '大正']
df_taisho.head(10)
````

````python
df_taisho.sort_values(by=['都道府県名', '人口（総数）'])
````

**ポイント**

- それぞれのカラムに対して、昇順と降順を別々に指定することができます。
- 引数`ascending`に`True`と`False`をリストで指定します。

````python
df_taisho.sort_values(by=['都道府県名', '人口（総数）'], ascending=[True, False])
````

## インデックスを並び替える | sort_index

`sort_index`メソッドを使用すると、インデックス名を並び替えることができます。

引数`ascending`のデフォルトは`True`（昇順）です。

dfのインデックスを降順に並び替えてみます。

````python
df.sort_index(ascending=False)
````

## カラムを並び替える | sort_index(axis=1)

`sort_index`メソッドの引数`axis`に「1」を指定すると、カラムを並び替えることができます。

引数`ascending`に`False`を指定すると降順になります。

dfのカラムを降順に並び替えてみます。

````python
df_sort_index = df.sort_index(axis=1, ascending=False)
df_sort_index
````

降順にするとデータの構造が分かりにくくなりますね。

## カラムを任意の順番に並び替える | DataFrame.reindex

カラムを任意の順番に並び替えをする場合は、`DataFrame.reindex`メソッドの引数`index`にリストで並び替えたカラムを指定します。

````python
df_sort_index.head()
````

````python
df_sort_index.reindex(columns = ['都道府県コード', '都道府県名', '元号', '和暦（年）', '西暦（年）', '人口（総数）', '人口（男）', '人口（女）'])
````

**ポイント**

インデックスを任意の順番に並び替える場合は、引数`index`に並び替えたインデックスをリストで指定します。
